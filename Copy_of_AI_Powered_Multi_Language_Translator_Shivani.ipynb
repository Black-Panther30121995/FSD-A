```python
from transformers import pipeline
from langdetect import detect
from gtts import gTTS
import gradio as gr
import os


translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ROMANCE")
```

```python
# Detect language of input
def detect_language(text):
    try:
        return detect(text)
    except:
        return "en"  # fallback

# Translate text based on detected language
def translate_text(text, target_lang):
    source_lang = detect_language(text)
    model_map = {
        ("en", "fr"): "Helsinki-NLP/opus-mt-en-fr",
        ("en", "de"): "Helsinki-NLP/opus-mt-en-de",
        ("en", "es"): "Helsinki-NLP/opus-mt-en-es",
        ("en", "hi"): "Helsinki-NLP/opus-mt-en-hi",
        ("fr", "en"): "Helsinki-NLP/opus-mt-fr-en",
        ("de", "en"): "Helsinki-NLP/opus-mt-de-en",
        ("es", "en"): "Helsinki-NLP/opus-mt-es-en",
        ("hi", "en"): "Helsinki-NLP/opus-mt-hi-en"
    }


    key = (source_lang, target_lang)
    model_name = model_map.get(key)
    if model_name is None:
        return f"Translation from {source_lang} to {target_lang} not supported."

    # Load specific model on-the-fly
    pipe = pipeline("translation", model=model_name)
    translated = pipe(text)
    return translated[0]['translation_text']
```

```python
def text_to_speech(text, lang_code='en'):
    tts = gTTS(text=text, lang=lang_code)
    tts.save("output.mp3")
    return "output.mp3"
```

```python
# Map display name to language code
language_code_map = {
    "French": "fr",
    "German": "de",
    "Spanish": "es",
    "Hindi": "hi"
}
```

```python
def full_translate(input_text, target_language_name):
    target_language_code = language_code_map[target_language_name]
    translation = translate_text(input_text, target_language_code)
    audio_file = text_to_speech(translation, target_language_code)
    return translation, audio_file

iface = gr.Interface(
    fn=full_translate,
    inputs=[
        gr.Textbox(label="Enter text to translate"),
        gr.Dropdown(
            choices=list(language_code_map.keys()),
            label="Target Language",
            value="French"
        )
    ],
    outputs=[
        gr.Textbox(label="Translated Text"),
        gr.Audio(label="Text-to-Speech Output")
    ],
    title="🌐 AI-Powered Multi-Language Translator",
    description="Enter text and choose target language. The app auto-detects the input language and returns a context-aware translation with audio."
)

iface.launch(debug=True)
```
